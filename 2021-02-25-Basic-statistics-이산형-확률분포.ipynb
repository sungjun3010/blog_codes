```python
import os
from scipy.stats import poisson
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
save_path = "./img/2021-02-25"
if not os.path.isdir(save_path):
    os.mkdir(save_path)
```

```python
times = list(range(30))
labels = []

f, ax = plt.subplots(1, 1)
for mu in [1,2,4,8,16]:
    labels.append(f'mu={mu}')
    sns.lineplot(x=times, y=poisson.pmf(times,mu=mu))
plt.legend(labels=labels)
plt.savefig(f"{save_path}/Poisson.png")
```
